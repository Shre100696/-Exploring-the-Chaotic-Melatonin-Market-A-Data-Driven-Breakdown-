# First Analysis : Proportion of each product in each category

```python
import matplotlib.pyplot as plt

num_kids = kids_final_data.shape[0]
num_adults = adults_final_data.shape[0]
total_products = num_kids + num_adults

kids_colors = ['pink', 'red']
adult_colors = ['blue', 'cornflowerblue', 'dodgerblue', 'deepskyblue', 'steelblue', 'royalblue']

kids_brands = kids_final_data.groupby('brand')
adult_brands = adults_final_data.groupby('brand')

# count of product in category-brand combination
kids_sizes = []
kids_labels = []
adult_sizes = []
adult_labels = []

for name, group in kids_brands:
    kids_sizes.append(group.shape[0])
    kids_labels.append(f"{name} ({group.shape[0]})")

for name, group in adult_brands:
    adult_sizes.append(group.shape[0])
    adult_labels.append(f"{name} ({group.shape[0]})")

explode_kids = [0.1 if i == 0 else 0 for i in range(len(kids_sizes))]
explode_adults = [0 if i != len(adult_sizes) - 1 else 0.1 for i in range(len(adult_sizes))]

# pie chart for Category and Brand
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
ax1.pie(kids_sizes, labels=kids_labels, colors=kids_colors, autopct='%1.1f%%', shadow=True, startangle=90, explode=explode_kids)
ax1.add_artist(plt.Circle((0, 0), 0.7, color='white'))
ax1.set_title(f"Proportion of Kids Products ({num_kids})")

ax2.pie(adult_sizes, labels=adult_labels, colors=adult_colors, autopct='%1.1f%%', shadow=True, startangle=90, explode=explode_adults)
ax2.add_artist(plt.Circle((0, 0), 0.7, color='white'))
ax2.set_title(f"Proportion of Adults Products ({num_adults})")

plt.show()
```

# Second Analysis :  Time Series Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt


df = adults_final_data

df['review_date'] = pd.to_datetime(df['review_date'])
df = df[df['review_date'].dt.year.between(2019, 2020)]


product_review_counts = df.groupby('product_title')['total_reviews'].sum().sort_values(ascending=False)

top_10_products = product_review_counts.head(10).index.tolist()
df_top_10_products = df[df['product_title'].isin(top_10_products)]
fig, ax = plt.subplots(figsize=(12, 6))
for product in top_10_products:
    df_product = df_top_10_products[df_top_10_products['product_title'] == product]
    df_product['review_date'].value_counts().sort_index().plot(ax=ax, label=product)
ax.set_xlabel('Review Date')
ax.set_ylabel('Number of Reviews')
ax.set_title('Review Volume Over Time for Top 10 Products (2019-2020)')
ax.legend()
plt.show()
```

# Forth Analysis : Success of each Brand based on review counts and average ratings

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = adults_final_data


brand_review_counts = df.groupby('brand')['total_reviews'].sum()


brand_avg_ratings = df.groupby('brand')['average_rating'].mean()


fig, ax = plt.subplots(1, 2, figsize=(12, 6))
sns.barplot(x=brand_review_counts, y=brand_review_counts.index, ax=ax[0])
sns.barplot(x=brand_avg_ratings, y=brand_avg_ratings.index, ax=ax[1])

ax[0].set_xlabel("Number of Reviews")
ax[0].set_ylabel("Brand")
ax[0].set_title("Review Volume by Brand")

ax[1].set_xlabel("Average Rating")
ax[1].set_ylabel("Brand")
ax[1].set_title("Average Rating by Brand")

plt.tight_layout()
plt.show()
```
